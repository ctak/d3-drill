```python
import pandas as pd
```

```python
df = pd.read_csv('cs.csv')
df
```

```python
df.drop(columns=['STEP'], inplace=True)
df
```

```python
df2 = df.set_index('Datetime').copy()
df2
```

```python
df2 = df2.stack(level=-1).reset_index().copy()
df2
```

```python
df2.rename(columns={"level_1": "division", 0: "val"}, inplace=True)
df2
```

```python
df2.to_csv('cs_stack.csv', index=False)
```
